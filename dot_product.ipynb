# The dot product

### Mohsin Javed, June 03, 2019

The dot product of two vectors $v$ and $w$ in $\mathbb{R}^3$ is defined as

$$v \cdot w = v_1 w_1 + v_2 w_2 + v_3 w_3.$$

The question is, why? Where does this definition come from? 

Also, we have a second definition which is usually motivated through 
the study of Physics:

$$v \cdot w = \|v\|\|w\|\cos \theta.$$

How are these two definitions equivalent?

### The algebraic definition
Let us pretend that we don't know the definition of the dot product. We first inspect the length of 
the vector $v$. This number is 

$$\|v\| = v_1^2 + v_2^2 + v_3^2,$$

where $v_1, v_2$ and $v_3$ are the components of the vector under a certain basis. This is 
just the might Pythagoras theorem. If we now change the basis, the vector $v$ might have a 
different representation 

$$v = [\hat{v}_1, \hat{v}_2, \hat{v}_3 ]^T.$$

But the length of the vector $v$ is invariant under a change of basis, and therefore, 

$$\hat{v}_1^2 + \hat{v}_2^2 + \hat{v}_3^2 = \|v\| = v_1^2 + v_2^2 + v_3^2.$$

Similarly, for $w$,

$$\hat{w}_1^2 + \hat{w}_2^2 + \hat{w}_3^2 = \|v\| = w_1^2 + w_2^2 + w_3^2.$$

The same must be true for the vector $u = v+w$. The length of $u$ must not 
change if we change the basis:

$$(\hat{v}_1+\hat{w}_1)^2 + (\hat{v}_2+\hat{w}_2)^2 + (\hat{v}_3+\hat{w}_3)^2 = \|u\| 
= (v_1+w_1)^2 + (v_2+w_2)^2 + (v_3+w_3)^2.$$

Expanding the above equation, we end up with

$$\|v\| + \|w\| + 2(\hat{v}_1\hat{w}_1 + \hat{v}_2\hat{w}_2 + \hat{v}_3\hat{w}_3)
= \|v\| + \|w\| + 2(v_1w_1 + v_2w_2 + v_3w_3),$$

and after cancellation, we arrive at another invariant quantity:

$$\hat{v}_1\hat{w}_1 + \hat{v}_2\hat{w}_2 + \hat{v}_3\hat{w}_3 = v_1w_1 + v_2w_2 + v_3w_3.$$

The above equation is telling us that the sum of the component-wise product 
of two vectors must not change if we change the basis. Therefore,
it makes sense to define a quantity called _dot product_ as 

$$v \cdot w = v_1 w_1 + v_2 w_2 + v_3 w_3.$$

Notice the left hand side of the equation is referring to the vectors $v$ and $w$ without any reference to a basis. The right hand side on the other hand tells that given a basis, compute the the sum of component-wise product, and as shown earlier, no matter what basis we choose, this number does not change. This explains the first definition, but what about the second definition?

### The Physicsy definition

We now derive the second definition form the first. Since the dot product is invariant under a change of basis, we are free to choose any basis. We choose a basis such that the x-axis of the new basis is parallel to $v$. (Imagine the xyz axes being rotated such that the x-axis is pointing in the same direction as $v$.) In this basis, $v$ has a simple representation

$$v = [\|v\|, 0, 0]^T.$$

What about $w$? If $w$ makes an angle $\theta$ with $v$, it makes the same angle $\theta$ with the new x-axis. Therefore, under the new basis, $w$ has the representation

$$w = [\|w\|\cos \theta, \hat{w}_2, \hat{w}_3]^T.$$

The invariant dot product under this basis is given by

\begin{align*}
v \cdot w &= \|v\| \|w\| \cos \theta + (0) \hat{w}_2 + (0) \hat{w}_2 \\
v \cdot w &=  \|v\| \|w\| \cos \theta.
\end{align*}

And we have arrive at the second definition of the dot product. The interesting aspect of the above equation is that both sides are completely independent of any choice of basis. 
The right hand side only depends on the lengths of the two vectors and the angle between them, all invariants!
